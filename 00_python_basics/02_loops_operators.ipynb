#### 1. Arithmetic Operators

```python
a = 10
b= 3

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
 

 # Comaprison Operator
print()
print(a>b)
print(a<b)
print(a==b)
print(a!=b)
```

#### 2. CONDITIONS (if / else)

```python
age = 20

if age >= 18:
    print("eligible for ai learning")
else:
    print("not eligible")
```

```python
marks = 75

if marks >= 80:
    print("Grade A")
elif marks >= 60:
    print("Grade B")

else:
    print("Grade C")
```

#### 3. Loops


```python
# for loop

for i in range(5):
    print("Learning AI", i)
```

```python
# while loop

count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
# Even or Odd Checker

number = int(input("enter a Number: "))

if number % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```
