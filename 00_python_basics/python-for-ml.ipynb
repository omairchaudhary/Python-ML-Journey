**CONDITIONS (IF / ELSE)**

```python
age = 20

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

**FUNCTIONS**

```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

```python
def mean(numbers):
    return sum(numbers) / len(numbers)

print(mean([10, 20, 30]))
```

```python
marks = [70, 85, 60, 90, 40]

def analyze(marks):
    avg = sum(marks) / len(marks)
    print("Average:", avg)

    if avg >= 50:
        print("Pass")
    else:
        print("Fail")

analyze(marks)
```
