```python
# Variables store values in memory
name = "Alex"   # string
age = 20         # integer
height = 5.9     # float
is_student = True  # boolean

# Display values
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student?", is_student)

# Check types
print(type(name), type(age), type(height), type(is_student))
```

```python
# Taking input from user
user_name = input("Enter your name: ")
user_age = int(input("Enter your age: "))
user_height = float(input("Enter your height in meters: "))

# Print with formatting
print(f"Hello {user_name}, you are {user_age} years old and {user_height}m tall.")
```

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)   # float
print("Floor division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    grade = 'A+'
elif marks >= 75:
    grade = 'A'
elif marks >= 60:
    grade = 'B'
elif marks >= 50:
    grade = 'C'
else:
    grade = 'F'

print(f"Your grade is: {grade}")
```

```python
# Print numbers from 1 to 10
for i in range(1, 11):
    print(i, end=" ")

# Sum of first 10 numbers
sum_total = 0
for i in range(1, 11):
    sum_total += i
print("\nSum:", sum_total)
```

```python
count = 1
while count <= 10:
    print(count, end=" ")
    count += 1
```

```python
# Break
for i in range(1, 11):
    if i == 5:
        break
    print(i, end=" ")

# Continue
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i, end=" ")  # prints only odd numbers
```

```python
# Function with parameters and return
def add_numbers(a, b):
    result = a + b
    return result

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
sum_result = add_numbers(x, y)
print("Sum:", sum_result)

# Function without parameters
def greet():
    print("Hello! Welcome to Python Learning.")

greet()
```

```python
# Creating a list
fruits = ["apple", "banana", "mango", "orange"]
print(fruits)

# Access elements
print(fruits[0])   # first
print(fruits[-1])  # last

# Add elements
fruits.append("grape")
fruits.insert(1, "kiwi")
print(fruits)

# Remove elements
fruits.pop()       # last element
fruits.remove("banana")
print(fruits)

# Loop through list
for fruit in fruits:
    print(fruit)
```

```python
student = {
    "name": "Umair",
    "age": 20,
    "marks": 85
}

# Access
print(student["name"])

# Add or update
student["grade"] = "A"
student["marks"] = 90

# Loop
for key, value in student.items():
    print(key, ":", value)
```
